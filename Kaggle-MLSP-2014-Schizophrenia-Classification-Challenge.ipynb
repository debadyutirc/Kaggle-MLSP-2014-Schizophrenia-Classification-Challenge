```python
# Helper Libs
import numpy as np
from scipy import stats
import pandas as pd
from time import time
import matplotlib.pyplot as plt

# ML Libs
from sklearn import svm
from sklearn.model_selection import GridSearchCV
from sklearn.decomposition import PCA
```

```python
# File paths
outputFile = 'mlsp2014Solution.csv' # Will be created at the end

# From the Kaggle Data
# Train Feature Files
trainFNCFile = 'Train/train_FNC.csv'
trainSBMFile = 'Train/train_SBM.csv'
trainLabelsFile = 'Train/train_labels.csv'

# Test Feature Files
testFNCFile = 'Test/test_FNC.csv'
testSBMFile = 'Test/test_SBM.csv'
```

```python
# Reading all the csv files into respective data frames for analysis
# Training data
trainFNC = pd.read_csv(trainFNCFile)
trainSBM = pd.read_csv(trainSBMFile)
Labels = pd.read_csv(trainLabelsFile)

# Testing data
testFNC = pd.read_csv(testFNCFile)
testSBM = pd.read_csv(testSBMFile)
```

```python
trainIDs = trainFNC['Id']
trainFNCAn = trainFNC.drop(['Id'], axis=1)
trainSBMAn = trainSBM.drop(['Id'], axis=1)

trainData = pd.concat([trainFNCAn, trainSBMAn], axis=1)
trainLabels = Labels.drop(['Id'], axis=1)
```

```python
trainData.head()
```

```python
testIDs = testFNC['Id']
testFNCAn = testFNC.drop(['Id'], axis=1)
testSBMAn = testSBM.drop(['Id'], axis=1)

testData = pd.concat([testIDs, testFNCAn, testSBMAn], axis=1)
testDataAn = pd.concat([testFNCAn, testSBMAn], axis=1)
```

```python
testDataAn.head()
```

```python
pcaTest = PCA()
pcaTest.fit(trainData)

plt.plot(np.cumsum(pcaTest.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
# Principal Component Analysis

t0 = time()
n_components = 70

pca = PCA(n_components, svd_solver='randomized', whiten=True)
pca.fit(trainData)

expRatio = 100 * pca.explained_variance_ratio_.sum()
print("Use of %d principal components explains %3.2f %% variance." % (n_components, expRatio))
print("PCA done in %0.3fs" % (time() - t0))
```

```python
trainDataTrx = pca.transform(trainData)
```

```python
t0 = time()
C_range = np.logspace(-1, 4, num=8)
gamma_range = np.logspace(-4, 1, num=8)
parameters = dict(gamma=gamma_range, C=C_range)
model = svm.SVC(kernel='rbf')
random_state = 1
gs = GridSearchCV(estimator=model, param_grid=parameters, cv=5, n_jobs=-1, verbose=1)
gs = gs.fit(trainDataTrx, trainLabels)

print(gs.best_params_)

print("Search for best parameters done in %0.3fs" % (time() - t0))
```

```python
t0 = time()
# C_value = 71.96856730011521
# gamma_value = 0.0005179474679231213

# svmClf = svm.SVC(C=C_value, gamma=gamma_value, kernel='rbf', probability=True)
svmClf = svm.SVC(probability=True)
svmClf.fit(trainDataTrx, trainLabels)

print("Training SVM done in %0.3fs" % (time() - t0))
```

```python
testDataTrx = pca.transform(testDataAn)
```

```python
random_state = 1
predictions = svmClf.predict_proba(testDataTrx)[:, 1]
```

```python
predictions.shape
```

```python
submission = {'ID' : testIDs, 
              'probability' : predictions}
submission = pd.DataFrame(submission)
submission.to_csv(outputFile, index = 0, float_format='%11.6f')
```
